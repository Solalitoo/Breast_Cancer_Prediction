# Classification Models 
Since our goal is to diagnose a patient's cancer, we are dealing with discrete outcomes (e.g: Benign:0 or Malignant:1).Thus classification models are more relevant. 

We will test several classification algorithms : 
- Logistic Regression 
- Decision Tree 
- Random Forest 
- K-Nearest Neighbours
- Suport Vector Machine 
- Naive Bayes 

Each model performance is assessed through specificty, sensitivity and accuracy 
- Specificty : Number of true Negative (%) 
  $$ \text{Specificity}= \frac{TN}{TN+FP}$$
- Sensitivity : Number of true Positive (%)
    $$ \text{Sensitivity}= \frac{TP}{TP+FN}$$
- Accuracy : Number of correct predictions (%)
  $$ \text{Accuracy}= \frac{TP+TN}{TP+NP+TN+FP}$$

## Splitting the data

```python
import sklearn 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


# Splitting the data 
X_train, X_val, y_train, y_val = train_test_split(
    breast_data.drop('diagnosis', axis=1), 
    breast_data['diagnosis'], 
    random_state=42
)

# Standardize features
ss = StandardScaler() #Standardize features by removing the mean and scaling to unit variance.
X_train = ss.fit_transform(X_train)
X_val = ss.fit_transform(X_val)
```

## Logistic Regression 

```python
import sklearn 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix



LR_model = LogisticRegression(max_iter=1000,random_state=0)
LR_model.fit(X_train, y_train)
y_pred_LR = LR_model.predict(X_val)
acc_LR = accuracy_score(y_val,y_pred_LR)
matrix=confusion_matrix(y_val,y_pred_LR)
print(f"{accuracy_score(y_val,y_pred_LR)}")
print(confusion_matrix(y_val,y_pred_LR))
print(f"Sensitivity : {(matrix[0,0])/(matrix[0,0]+matrix[1,0])}")
print(f"Specificity : {(matrix[1,1])/(matrix[1,0]+matrix[1,1])}")
```

## Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
#Specify and fit the model 
tree_model2 = DecisionTreeClassifier(random_state=1)
tree_model2.fit(X_train,y_train)
y_pred_tree = tree_model2.predict(X_val)
matrix= confusion_matrix(y_val,y_pred_tree)
acc_tree = accuracy_score(y_val,y_pred_tree)
print(f"Accuracy : {accuracy_score(y_val,y_pred_tree)} ")
print(f"Sensitivity : {(matrix[0,0])/(matrix[0,0]+matrix[1,0])}")
print(f"Specificity : {(matrix[1,1])/(matrix[1,0]+matrix[1,1])}")
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train,y_train)
y_pred_rf = rf_model.predict(X_val)
matrix= confusion_matrix(y_val,y_pred_rf)
acc_rf = accuracy_score(y_val,y_pred_rf)
print(f"Accuracy : {accuracy_score(y_val,y_pred_rf)} ")
print(f"Sensitivity : {(matrix[0,0])/(matrix[0,0]+matrix[1,0])}")
print(f"Specificity : {(matrix[1,1])/(matrix[1,0]+matrix[1,1])}")
```

## K_ Nearest Neighbors 

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
A =[]

# find the optimum value for k 
for k in range (1,42):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train,y_train)
    y_pred_knn = knn_model.predict(X_val)
    A.append(np.mean(y_val !=y_pred_knn))

plt.figure(figsize=(14,6))
plt.plot(range(1,42), A,color='purple', linestyle="--",
         marker='o', markersize=10, markerfacecolor='b')

plt.title('Error Rate vs K-value')
plt.show
```

```python
k=35
knn_model = KNeighborsClassifier(n_neighbors=k)
knn_model.fit(X_train,y_train)
y_pred_knn = knn_model.predict(X_val)
matrix= confusion_matrix(y_val,y_pred_knn)
acc_knn = accuracy_score(y_val,y_pred_knn)
print(f"Accuracy : {accuracy_score(y_val,y_pred_knn)} ")
print(f"Sensitivity : {(matrix[0,0])/(matrix[0,0]+matrix[1,0])}")
print(f"Specificity : {(matrix[1,1])/(matrix[1,0]+matrix[1,1])}")
```
